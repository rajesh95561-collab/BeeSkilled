```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tdf = pd.read_csv('Titanic-Dataset.csv')
tdf
```

```python
tdf.info()
```

```python
tdf['PassengerId'] = tdf['PassengerId'].astype(np.int8)
tdf['Survived'] = tdf['Survived'].astype(np.int8)
tdf['Pclass'] = tdf['Pclass'].astype(np.int8)
tdf['SibSp'] = tdf['SibSp'].astype(np.int8)
tdf['Parch'] = tdf['Parch'].astype(np.int8)
```

```python
tdf['Age'] = tdf['Age'].fillna(tdf['Age'].median())
```

```python
tdf.info()
```

```python
tdf[tdf['Cabin'].isnull()]
```

```python
tdf = tdf.drop(columns= ['Cabin'])
```

```python
tdf[tdf['Embarked'].isnull()]
tdf['Embarked'] = tdf['Embarked'].ffill()
```

```python
tdf.info()
```

```python
tdf.describe()
```

```python
tdf['Sex'].unique()
tdf['Sex'] = tdf['Sex'].map({'male':1,'female':0})
```

```python
tdf['Sex'].values
```

```python
tdf['Embarked'].unique()
tdf['Embarked'] = tdf['Embarked'].map({'C':1, 'Q':2, 'S':3})
```

```python
tdf['Embarked'].values
```

```python
tdf
```

```python
#visualize of age distribution
plt.figure(figsize=(8,5))
sns.histplot(tdf['Age'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
tdf.to_csv('cleaned_titanic_dataset.csv',index=False)
print('Data Preprocessing Complete and Cleaned file save sucessfully')
```
